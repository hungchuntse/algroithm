```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
import functools
import operator
import random
from random import shuffle
import time


class final_project():
    def __init__(self,tbatch, week1, week2,week3, week4,count2,count3,count4,total_1,total_2,total_3,total_4, List_1,List_2,List_3,List_4, week1_1,week2_1,week3_1,week4_1, month_1,month_2,month_3,month_4, t_week_1,t_week_2,t_week_3,t_week_4, t_month_1,t_month_2,t_month_3,t_month_4,count,tls,total_tls,week_a,week_ac,week_as,df_week1,df_week2,df_week3,df_week4):
        #整合20為一群的資料
        self.total_1 = total_1
        self.total_2 = total_2
        self.total_3 = total_3
        self.total_4 = total_4
        #分20為一群的資料
        self.List_1 = List_1
        self.List_2 = List_2
        self.List_3 = List_3
        self.List_4 = List_4
        #分該週的週檢
        self.cweek1 = week1_1
        self.cweek2 = week2_1
        self.cweek3 = week3_1
        self.cweek4 = week4_1
        #分該週的月檢
        self.cmonth1 = month_1
        self.cmonth2 = month_2
        self.cmonth3 = month_3
        self.cmonth4 = month_4
        #整合第一週的週檢
        self.t_week_1 = t_week_1
        self.t_week_2 = t_week_2
        self.t_week_3 = t_week_3
        self.t_week_4 = t_week_4
        #整合第一週的月檢
        self.t_month_1 = t_month_1
        self.t_month_2 = t_month_2
        self.t_month_3 = t_month_3
        self.t_month_4 = t_month_4
        
        #累加一天已用的 TLS (num)
        self.tls=tls
        #整合一週的 self.tls (list)
        self.total_tls=total_tls
        #整合self.t_week_1與self.t_month_1
        self.count=count
        #整合self.t_week_i與self.t_month_i及前一週剩下的
        self.count2=count2
        self.count3=count3
        self.count4=count4
        #整合剩餘的紙管(list)
        self.week_a=week_a
        #剩餘的TLS數量
        self.week_as=week_as
        #剩餘多少種產品(count)
        self.week_ac=week_ac
        
        self.tbatch=tbatch
        
       
        self.df_week1=df_week1
        self.df_week2=df_week2
        self.df_week3=df_week3
        self.df_week4=df_week4
        
        self.week1 = week1
        self.week2 = week2
        self.week3 = week3
        self.week4 = week4
    
    def csv(self):
        self.df_week1=pd.read_csv("week1.csv",encoding="big5")
        self.df_week2=pd.read_csv("week2.csv",encoding="big5")
        self.df_week3=pd.read_csv("week3.csv",encoding="big5")
        self.df_week4=pd.read_csv("week4.csv",encoding="big5")
        
        self.week1=self.df_week1.values.tolist()
        self.week2=self.df_week2.values.tolist()
        self.week3=self.df_week3.values.tolist()
        self.week4=self.df_week4.values.tolist()
    
    def classification(self, week, total, List):
        for gen in range(70, 80):
            low = gen
            for j in range(21):
                for i in range(len(week)):
                    if int(week[i][0] / 10000) == low:
                        List.append(week[i])
                low += 1
            total.append(List)
            List = []
            

    def classification_c2t(self, total, cweek, cmonth, t_week, t_month):
        for i in range(len(total)):
            for j in range(len(total[i])):
                cal = total[i][j][0] % 100
                if cal == 68 or cal == 42 or cal == 0:
                    cweek.append(total[i][j])
                else:
                    cmonth.append(total[i][j])
            
            t_week.append(cweek)
            t_month.append(cmonth)
            cweek = []
            cmonth = []

    def sort(self):
        for i in range(len(self.t_month_1)):
            np.random.shuffle(self.t_month_1[i])
        for i in range(len(self.t_month_2)):
            np.random.shuffle(self.t_month_2[i])
        for i in range(len(self.t_month_3)):
            np.random.shuffle(self.t_month_3[i])
        for i in range(len(self.t_month_4)):
            np.random.shuffle(self.t_month_4[i])
        for i in range(len(self.t_week_1)):
            np.random.shuffle(self.t_week_1[i])
        for i in range(len(self.t_week_2)):
            np.random.shuffle(self.t_week_2[i])
        for i in range(len(self.t_week_3)):
            np.random.shuffle(self.t_week_3[i])
        for i in range(len(self.t_week_4)):
            np.random.shuffle(self.t_week_4[i])
            
    
    def batch1(self,t_week_1,t_month_1):
        self.count.append(self.t_week_1)
        self.count.append(self.t_month_1)
        for k in range(5):
            for i in range (len(self.count)):
                for l in range(len(self.count[i])):
                    for j in range(len(self.count[i][l])):
                        self.tls=self.tls+self.count[i][l][j][1]

                        if self.tls<=672:
                           
                           
                            
                            e=l
                            
                            while e <len(self.count[i])-1:
                                e=e+1


                                for m in range(len(self.count[i][e])):
                                    if self.count[i][l][j][2] == self.count[i][e][m][2]:
                                        self.count[i][e][m][0]=0
                                        self.count[i][e][m][1]=0
                                        self.count[i][e][m][2]=0
                                        self.count[i][e][m][3]=0

                            
                            self.count[i][l][j][0]=0
                            self.count[i][l][j][1]=0
                            self.count[i][l][j][2]=0
                            self.count[i][l][j][3]=0
                            

                        if self.tls>672:
                            self.tls=self.tls-self.count[i][l][j][1]
                            
                                

                            

            self.total_tls.append(self.tls)
            self.tls=0
        self.tbatch.append(self.total_tls)
            
            
        
#         print("batch1",self.total_tls)
#         print("----------------------")
        
  
    def batch2(self,t_week_2,t_month_2):
        self.total_tls=[]
        self.count2.append(self.t_week_2)
        self.count2.append(self.count[1])
        self.count2.append(self.t_month_2)

        for k in range(5):
            for i in range (len(self.count2)):
                for l in range(len(self.count2[i])):
                    for j in range(len(self.count2[i][l])):
                        self.tls=self.tls+self.count2[i][l][j][1]

                        if self.tls<=672:
                           
                           
                            
                            e=l
                            
                            while e <len(self.count2[i])-1:
                                e=e+1


                                for m in range(len(self.count2[i][e])):
                                    if self.count2[i][l][j][2] == self.count2[i][e][m][2]:
                                        self.count2[i][e][m][0]=0
                                        self.count2[i][e][m][1]=0
                                        self.count2[i][e][m][2]=0
                                        self.count2[i][e][m][3]=0

                            
                            self.count2[i][l][j][0]=0
                            self.count2[i][l][j][1]=0
                            self.count2[i][l][j][2]=0
                            self.count2[i][l][j][3]=0
                            
                        if self.tls>672:
                            self.tls=self.tls-self.count2[i][l][j][1]
                            
                                

                            

            self.total_tls.append(self.tls)
            self.tls=0
        self.tbatch.append(self.total_tls)
           
            
        
#         print("batch2",self.total_tls)
#         print("----------------------")
        
        
    def batch34(self,tcount,t_week,lcount,t_month):
        self.total_tls=[]
        tcount.append(t_week)
        tcount.append(lcount[2])
        tcount.append(t_month)
    
        for k in range(5):
            for i in range (len(tcount)):
                for l in range(len(tcount[i])):
                    for j in range(len(tcount[i][l])):
                        self.tls=self.tls+tcount[i][l][j][1]

                        if self.tls<=672:
                           
                            
                            
                            e=l
                            
                            while e <len(tcount[i])-1:
                                e=e+1


                                for m in range(len(tcount[i][e])):
                                    if tcount[i][l][j][2] == tcount[i][e][m][2]:
                                        tcount[i][e][m][0]=0
                                        tcount[i][e][m][1]=0
                                        tcount[i][e][m][2]=0
                                        tcount[i][e][m][3]=0

                            
                            tcount[i][l][j][0]=0
                            tcount[i][l][j][1]=0
                            tcount[i][l][j][2]=0
                            tcount[i][l][j][3]=0
                            

                        if self.tls>672:
                            self.tls=self.tls-tcount[i][l][j][1]
                            
                                

                            

            self.total_tls.append(self.tls)
            self.tls=0
        self.tbatch.append(self.total_tls)
            
            
            
        
#         print("batch3",self.total_tls)
#         print("----------------------")
#         print(self.tbatch)
        
       
    def batch1234_result(self,t_count):
        self.week_a=[]
        self.week_ac=0
        self.week_as=0
        for i in range(len(t_count)):
            for l in range(len(t_count[i])):
                for j in range(len(t_count[i][l])):
                    if t_count[i][l][j][3] !=0:                
                        self.week_a.append(t_count[i][l][j])
                        t_count[i][l][j][3]=0
                        self.week_as=self.week_as+t_count[i][l][j][1]
            
        for k in range(len(self.week_a)):
            self.week_a[k][3]=1  
        print('---------------------------------')    
        print('剩餘',self.week_a)
        print('剩餘數',self.week_as)

        self.week_ac=len(self.week_a)
#         print('---------------------------------')
#         print(self.week_ac)
    
    
    def main(self):
        
        self.csv()
        self.classification(self.week1, self.total_1, self.List_1)
        self.classification(self.week2, self.total_2, self.List_2)
        self.classification(self.week3, self.total_3, self.List_3)
        self.classification(self.week4, self.total_4, self.List_4)
        self.classification_c2t(self.total_1, self.cweek1, self.cmonth1, self.t_week_1, self.t_month_1)
        self.classification_c2t(self.total_2, self.cweek2, self.cmonth2, self.t_week_2, self.t_month_2)
        self.classification_c2t(self.total_3, self.cweek3, self.cmonth3, self.t_week_3, self.t_month_3)
        self.classification_c2t(self.total_4, self.cweek4, self.cmonth4, self.t_week_4, self.t_month_4)
        self.sort()
        
        self.batch1(self.t_month_1,self.t_week_1)
        self.batch1234_result(self.count)
        
        self.batch2(self.t_month_2,self.t_week_2)
        self.batch1234_result(self.count2)
        
        self.batch34( self.count3,self.t_week_3,self.count2, self.t_month_3)
        self.batch1234_result(self.count3)
        
        self.batch34( self.count4,self.t_week_4,self.count3, self.t_month_4)
        self.batch1234_result(self.count4)

if __name__ =="__main__":

    fin = final_project([],[], [], [], [], [],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],0,[],[],0,0,[],[],[],[])
    fin.main()
    
    
```

```python
tStart = time.time()
class SA():
    def __init__(self,T,Tmin,Tk,alpha,N,K,init_sol,init_remainder,init_obj,init_tc,final_sol,final_remainder,final_obj,final_tc,pltobj):
        self.T = T  #initiate temperature
        self.Tmin = Tmin  #minimum value of terperature
        self.Tk = Tk
        self.alpha=alpha
        self.N=N
        self.K=K
        
        self.init_sol=init_sol
        self.init_remainder=init_remainder
        self.init_obj=init_obj
        self.init_tc=init_tc
        
        self.final_sol=final_sol
        self.final_remainder=final_remainder
        self.final_obj=final_obj
        self.final_tc=final_tc
        
        self.pltobj=pltobj
        
    def run_sa(self):
        self.final_sol = self.init_sol
        self.final_remainder = self.init_remainder
        self.final_obj = self.init_obj
        self.final_tc = self.init_tc
        self.TK=self.T
        print("init sol：",self.init_sol)
        print("----------------------------") 
        print("init remainder：",self.init_remainder)
        print("----------------------------") 
        print("init obj：",self.init_obj)
        print("----------------------------") 
        print("init tc：",self.init_tc)
        print("----------------------------") 
        while self.Tk > self.Tmin:
            for k in range(self.K):
                for n in range(self.N):
                    fin = final_project([],[],[] , [], [], [],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],0,[],[],0,0,[],[],[],[])
                    fin.main()
                    new_sol = fin.tbatch
                    new_remainder = fin.week_a
                    new_obj = fin.week_as
                    new_tc =fin.week_ac
                    
                    if  new_obj <= self.init_obj:
                        self.init_sol = new_sol
                        self.init_remainder = new_remainder
                        self.init_obj = new_obj
                        self.init_tc = new_tc
                        
                       
                    else:
                        p = np.exp(-((new_obj - self.init_obj)/self.Tk))
                        if np.random.rand() <= p:
                            self.init_sol = new_sol
                            self.init_remainder = new_remainder
                            self.init_obj = new_obj
                            self.init_tc = new_tc
                            
                           
                           
                        else:
                            self.init_sol = self.init_sol
                            self.init_remainder = self.init_remainder
                            self.init_obj = self.init_obj
                            self.init_tc = self.init_tc
                            
                    if self.init_obj <= self.final_obj:
                        self.final_sol = self.init_sol
                        self.final_remainder = self.init_remainder
                        self.final_obj = self.init_obj
                        self.final_tc = self.init_tc
                        self.pltobj.append(self.final_obj)

            self.Tk = self.alpha * self.Tk
        print("Final sol：",self.final_sol)
        print("----------------------------") 
        print("Final remainder：",self.final_remainder)
        print("----------------------------") 
        print("Final obj：",self.final_obj)
        print("----------------------------") 
        print("Final tc：",self.final_tc)
        print("----------------------------")
        
        plt.figure()
        plt.title("Figure")
        plt.xlabel("number of iterations", size = 14)
        plt.ylabel("fitness", size = 14 )
        t = [t for t in range(self.K)]
        plt.plot(self.pltobj, color = 'b', linewidth = 2)
        plt.show()
        
    def main(self):
        self.run_sa()   
        
if __name__ =="__main__":

    sa = SA(1000,10,1000,0.5,10,10,fin.tbatch,fin.week_a,fin.week_as,fin.week_ac,[],[],0,0,[])
    sa.main()
    
tEnd = time.time()
print ("It cost %f sec" % (tEnd - tStart))
print (tEnd - tStart)
```
