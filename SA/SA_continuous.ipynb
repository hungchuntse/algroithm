```python
import math
import numpy as np

def function(x,y):
    value=(x ** 2) + (2 * y ** 2) - (0.3 * math.cos(3 * math.pi * x)) * (math.cos(4 * math.pi * y)) + 0.3
    return value

initx = np.random.uniform(-100,100)
inity = np.random.uniform(-100,100)
xf=initx
yf=inity
print ("x=",initx,"y=",inity)


initz=(initx**2)+(2*inity **2)-(0.3*math.cos(3*math.pi*initx )* (math.cos(4 * math.pi * inity )) + 0.3)


initT = 500 #初始温度
minT = 1 #温度下限
iterL = 100 #每个T值的迭代次数
delta = 0.5 #温度衰减系数
K = 10

nowT=initT

while nowT>minT:
    for k in np.arange(1,K,1):
        
        for i in np.arange(1,iterL,1):
            initz=function(initx,inity)
            new_initx=initx+(np.random.rand()-0.5)
            new_inity=inity+(np.random.rand()-0.5)
#             new_initx=initx+(2*np.random.rand()-1)
#             new_inity=inity+(2*np.random.rand()-1)
#             new_initx=initx*0.5
#             new_inity=inity*0.5
            
            a=function(new_initx,new_inity)-initz
            
            if function(new_initx,new_inity)<=initz:
                initx=new_initx #Xn+1
                inity=new_inity
                
            else:
                p = np.exp(-a/(nowT))
                if np.random.rand()<=p:
                    initx=new_initx 
                    inity=new_inity
                else:
                    initx=initx
                    inity=inity
            if function(initx,inity)<=function(xf,yf):
            
                xf=initx
                yf=inity
           # print(initx, inity)
            if iterL % 100 == 0:
                print("local",function(initx,inity))    
            
    nowT = nowT * delta
    
        
print("最佳解：","X=",xf,"Y=",yf)
print("最佳值：",function(xf,yf))
```
